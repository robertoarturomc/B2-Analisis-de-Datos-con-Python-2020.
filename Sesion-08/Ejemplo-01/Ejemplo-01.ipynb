## Ejemplo 01: Programaci칩n de clusterizaci칩n con K-Medias
En este ejemplo generaremos un algoritmo de clusterizaci칩n con K-medias. Considera que el algoritmo dar치 resultados aleatorios cada vez que lo ejecutes, sin embargo el proceso es el mismo.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
x,y = make_blobs(n_samples = 1000, centers = 9, cluster_std=1.5)
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
k_medias = KMeans(n_clusters = 9, max_iter = 300, init = 'random')
k_medias.fit(x)
plt.scatter(x[:,0], x[:,1])
plt.scatter(k_medias.cluster_centers_[:,0], k_medias.cluster_centers_[:,1], s=100, c='yellow')
plt.show()
```
