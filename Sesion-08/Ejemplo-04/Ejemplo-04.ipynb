## Ejemplo 04.
En este ejemplo vamos a catalogar 4 centros (o 4 valores diferentes) con un perceptr√≥n multicapa. Para ello, veamos los siguientes elementos, vamos crendo blobs de 4 clases artificialmente...

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
```

```python
x, y = make_blobs(n_samples=1000, centers=4, n_features=2)
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

Y hacemos la partici√≥n en entrenamiento y prueba como la vez pasada

```python
x_entrenamiento, x_prueba, y_entrenamiento, y_prueba = train_test_split(x, y, test_size=0.30)
plt.scatter(x_prueba[:,0], x_prueba[:,1], c=y_prueba)
```

```python
plt.scatter(x_entrenamiento[:,0], x_entrenamiento[:,1], c=y_entrenamiento)
```

## Entrenamiento.
Nota que en el entrenamiento estamos utilizando 4 neuronas de salida, esto se le conoce en la literatura de redes neuronales como 'one-hot' y hace que cada neurona sirva como un bit que mappea una salida. Podr√≠as usar dos neuronas para generar 4 estados ( 00, 01, 10, y 11) pero esto por lo general no da los mejores resulados de clasificaci√≥n...

```python
from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier(activation = 'logistic',solver='adam', alpha=1e-4, max_iter=10000, hidden_layer_sizes=(16,4))
classifier.fit(x_entrenamiento, y_entrenamiento)  
```

```python
resultados = classifier.predict(x_prueba)
errores = np.sum(np.abs(resultados - y_prueba))
accuracy = (len(y_prueba)-errores)/len(y_prueba) *100
print("Errores cometidos: "+str(errores))
print("Precisi√≥n: "+str(accuracy)+"%")
```

```python
print("Clases esperadas:")
print(y_prueba)
print("Resultados obtenidos:")
print(resultados)
print("Desfase:")
print(y_prueba-resultados)
```

## Sub-muestreo neuronal.
Ahora ve que pasa si solamente uso una neurona de salida en lugar de 4...

```python
classifier = MLPClassifier(activation = 'logistic',solver='adam', alpha=1e-4, max_iter=10000,
                           hidden_layer_sizes=(16,1))
classifier.fit(x_entrenamiento, y_entrenamiento) 
resultados = classifier.predict(x_prueba)
errores = np.sum(np.abs(resultados - y_prueba))
accuracy = (len(y_prueba)-errores)/len(y_prueba) *100
print("Errores cometidos: "+str(errores))
print("Precisi√≥n: "+str(accuracy)+"%")
print("Clases esperadas:")
print(y_prueba)
print("Resultados obtenidos:")
print(resultados)
print("Desfase:")
print(y_prueba-resultados)
```
