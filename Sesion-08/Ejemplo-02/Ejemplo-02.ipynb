## Ejemplo 02: Separación de datos con Numpy.
En este ejemplo vamos a generar blobs y vamos a separar los datos en entrenamiento y prueba, lo cual nos será de gran utilidad para el entrenamiento supervisado. Primero vamos a crear 1000 muestras, en 2 centros, con 2 características.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
```

```python
x, y = make_blobs(n_samples=1000, centers=2, n_features=2)
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

En SKLearn existe la función train_test_split, lo cual separará nuestros datos de forma aleatoria en elementos de entrenamiento y prueba. Vamos a graficar ambas muestras, y comparalas con los datos graficados en un principio, para que todos los datos relevantes se noten

```python
x_entrenamiento, x_prueba, y_entrenamiento, y_prueba = train_test_split(x, y, test_size=0.30)
plt.scatter(x_prueba[:,0], x_prueba[:,1], c=y_prueba)
```

```python
plt.scatter(x_entrenamiento[:,0], x_entrenamiento[:,1], c=y_entrenamiento)
```
